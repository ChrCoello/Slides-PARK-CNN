```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=[7, 7])
random_image = np.random.randint(low=0, high=255, size=(224,224))
plt.imshow(random_image,interpolation='nearest',cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
plt.axis('off')
plt.tight_layout()
plt.savefig('images/random_image_02.png')
```
