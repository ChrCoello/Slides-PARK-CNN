```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
random_image = np.random.randint(low=0, high=255, size=[224,224])
plt.imshow(random_image,interpolation='nearest',cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
```

```python
random_image = np.random.randint(low=0, high=255, size=[224,224,3])
plt.imshow(random_image,interpolation='nearest', vmin=0, vmax=255)
```
